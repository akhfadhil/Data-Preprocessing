```python
import pandas as pd
```

```python
df = pd.read_excel('D:\Dev\Tetris\Data Merge\TOKOPEDIA_SAMPLE 1.xlsx')
```

```python
df.columns
```

```python
columns = ['Invoice',  
           'Product Name', 
           'Product Price (IDR)', 
           'Product Quantity', 
           'Shipping Fee (IDR)', 
           'Total Amount (IDR)',
           'Order Created Date', 
           'Courier Name', 
           'Order Status', 
           'Region/Province']
```

```python
df.rename(columns = {'Payment Date' : 'Order Created Date', 
                     'Price (Rp.)' : 'Product Price (IDR)', 
                     'Quantity' : 'Product Quantity', 
                     'Total Shipping Fee (Rp.)' : 'Shipping Fee (IDR)', 
                     'Total Amount (Rp.)' : 'Total Amount (IDR)', 
                     'Courier' : 'Courier Name', 
                     'Recipient Address': 'Region/Province'}, 
                     inplace = True) 
```

```python
for i in df.columns:
    if i not in columns:
        df = df.drop([i], axis=1)
```

```python
df = df[columns]
```

```python
df
```

```python
# Order Status
df['Order Status'] = df['Order Status'].apply(lambda x: 'Order Berhasil' if 'selesai' in x else x)
df['Order Status'] = df['Order Status'].apply(lambda x: 'Order Gagal' if 'tidak valid' in x else x)
```

```python
df['Order Status'] = df['Order Status'].apply(lambda x: 'Order On Process' if 'Menunggu' in x else x)
df['Order Status'] = df['Order Status'].apply(lambda x: 'Order On Process' if 'sanan' in x else x)
```

```python
def category(row):
   if 'Baju' in row['Product Name']:
      return 'Pakaian'
   if 'Pashmina' in row['Product Name']:
      return 'Pakaian'
   if 'Sweater' in row['Product Name']:
      return 'Pakaian'
   if 'Jas' in row['Product Name']:
      return 'Pakaian'
   if 'Seragam' in row['Product Name']:
      return 'Pakaian'
   if 'Batik' in row['Product Name']:
      return 'Pakaian'
   if 'Sepatu' in row['Product Name']:
      return 'Sepatu'
   return 'Accessories'
```

```python
# Product Category
df['Product Category'] = df.apply(category, axis=1)
```

```python
# Courier Service
df['Courier Service'] = df['Courier Name'].apply(lambda x: 'Same Day' if 'Same' in x else 'Regular')
```

```python
# Courier Name

def courier_name(row):
   if 'GO-JEK' in row['Courier Name']:
      return 'GO-JEK'
   if 'JNE' in row['Courier Name']:
      return 'JNE'
   if 'SiCepat' in row['Courier Name']:
      return 'SiCepat'
   
df['Courier Name'] = df.apply(courier_name, axis=1)
```

```python
# ECommerce
df['ECommerce'] = 'Tokopedia'
```

```python
# Extract Province
df['Region/Province'] = df['Region/Province'].str.extract(r'(\D+$)')
```

```python
# Date Format
df['Date']= df['Order Created Date']
df['Date'] = pd.to_datetime(df.Date, format='%d-%m-%Y %H:%M:%S')
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
df = df.drop(['Order Created Date'], axis=1)
df.rename(columns = {'Date':'Order Created Date'}, inplace =True)
```

```python
new_cols_order = ['Invoice', 'Product Name', 'Product Category', 'Product Price (IDR)', 'Product Quantity',
       'Shipping Fee (IDR)', 'Total Amount (IDR)',  'Order Created Date', 'Courier Name', 'Courier Service',
       'Order Status', 'Region/Province', 'ECommerce']
```

```python
df = df[new_cols_order]
```
