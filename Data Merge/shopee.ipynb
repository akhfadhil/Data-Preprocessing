```python
import pandas as pd
```

SHOPEE

```python
df = pd.read_excel('D:\Dev\Tetris\Data Merge\SHOPEE_SAMPLE 1.xlsx')
```

```python
df.columns
```

```python
columns = ['Invoice',  
           'Product Name', 
           'Product Price (IDR)', 
           'Product Quantity', 
           'Shipping Fee (IDR)', 
           'Total Amount (IDR)',
           'Order Created Date', 
           'Courier Name', 
           'Order Status', 
           'Region/Province']
```

```python
df.rename(columns = {'No. Pesanan':'Invoice',
                     'Status Pesanan':'Order Status', 
                     'Opsi Pengiriman':'Courier Name',
                     'Waktu Pembayaran Dilakukan':'Order Created Date',
                     'Nama Produk':'Product Name', 
                     'Harga Awal':'Product Price (IDR)',
                     'Jumlah':'Product Quantity', 
                     'Ongkos Kirim Dibayar oleh Pembeli':'Shipping Fee (IDR)',
                     'Total Pembayaran':'Total Amount (IDR)', 
                     'Provinsi':'Region/Province'}, 
                     inplace = True) 
```

```python
for i in df.columns:
    if i not in columns:
        df = df.drop([i], axis=1)
```

```python
df = df[columns]
```

```python
df
```

```python
# Order Status
df['Order Status'] = df['Order Status'].apply(lambda x: 'Order Berhasil' if x=='Selesai' else 'Order Gagal')
```

```python
def category(row):
   if 'Baju' in row['Product Name']:
      return 'Pakaian'
   if 'Pashmina' in row['Product Name']:
      return 'Pakaian'
   if 'Sweater' in row['Product Name']:
      return 'Pakaian'
   if 'Jas' in row['Product Name']:
      return 'Pakaian'
   if 'Seragam' in row['Product Name']:
      return 'Pakaian'
   if 'Batik' in row['Product Name']:
      return 'Pakaian'
   if 'Sepatu' in row['Product Name']:
      return 'Sepatu'
   return 'Accessories'
```

```python
# Product Category
df['Product Category'] = df.apply(category, axis=1)
```

```python
# Courier Service
df['Courier Service'] = df['Courier Name'].apply(lambda x: 'Same Day' if 'Same' in x else 'Regular')
```

```python
# Courier Name

def courier_name(row):
   if 'Go' in row['Courier Name']:
      return 'GO-JEK'
   if 'JNE' in row['Courier Name']:
      return 'JNE'
   if 'SiCepat' in row['Courier Name']:
      return 'SiCepat'
   if 'Grab' in row['Courier Name']:
      return 'Grab'
   if 'J&T' in row['Courier Name']:
      return 'J&T'
   
df['Courier Name'] = df.apply(courier_name, axis=1)
```

```python
# Ecommerce
df['ECommerce'] = 'Shopee'
```

```python
# Extract Province
df['Region/Province'] = df['Region/Province'].str.extract(r"([^(Prop.\s)](\w+).+)")[0]
df['Region/Province'] = df['Region/Province'].apply(lambda x: 'Papua' if 'apua' in x else x)
```

```python
# Date Format
df['Date']= df['Order Created Date']
df['Date'] = pd.to_datetime(df.Date, format='%Y-%m-%d %H:%M')
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
df = df.drop(['Order Created Date'], axis=1)
df.rename(columns = {'Date':'Order Created Date'}, inplace =True)
```

```python
new_cols_order = ['Invoice', 'Product Name', 'Product Category', 'Product Price (IDR)', 'Product Quantity',
       'Shipping Fee (IDR)', 'Total Amount (IDR)',  'Order Created Date', 'Courier Name', 'Courier Service',
       'Order Status', 'Region/Province', 'ECommerce']
```

```python
df = df[new_cols_order]
```
